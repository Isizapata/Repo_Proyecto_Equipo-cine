```python
# Gráfico 1 Géneros nominados vs ganadores a "Mejor Película"

import pandas as pd
import altair as alt

from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Base de datos Entrega 3.csv", sep=";")
df.head()

comparacion = pd.crosstab(df['GÉNERO'], df['LUGAR'])
comparacion = comparacion.reset_index()

comparacion = comparacion.sort_values('Ganadora', ascending=False).head(10)

comparacion_long = comparacion.melt(
    id_vars='GÉNERO',
    var_name='Resultado',
    value_name='Cantidad'
)

grafico_1 = alt.Chart(comparacion_long).mark_bar().encode(
    x=alt.X('GÉNERO:N', sort='-y', title='Género'),
    y=alt.Y('Cantidad:Q', title='Cantidad'),
    color=alt.Color('Resultado:N', title='Resultado'),
    xOffset='Resultado:N',
    tooltip=['GÉNERO', 'Resultado', 'Cantidad']
).properties(
    title='Nominadas vs Ganadoras por Género (Top 10)',
    width=600,
    height=400
)

grafico_1
```

```python
grafico_1.save("grafico_1.html")
files.download("grafico_1.html")
```

```python
# Gráfico 2: Géneros nominados a "Mejor Película" a lo largo de los años

import pandas as pd
import altair as alt

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("Base de datos Entrega 3.csv", sep=";")

# Crear variable década
df['Década'] = (df['AÑO ESTRENO'] // 10) * 10

# Tabla de frecuencias por década y género
generos_decada = pd.crosstab(
    df['Década'],
    df['GÉNERO']
).reset_index()

# Pasar a formato largo
generos_decada_long = generos_decada.melt(
    id_vars='Década',
    var_name='Género',
    value_name='Cantidad'
)

# Opcional: eliminar combinaciones con 0 películas
generos_decada_long = generos_decada_long[
    generos_decada_long['Cantidad'] > 0
]

# Gráfico
grafico_2 = alt.Chart(generos_decada_long).mark_area().encode(
    x=alt.X('Década:O', title='Década'),
    y=alt.Y('Cantidad:Q', title='Cantidad de películas'),
    color=alt.Color('Género:N', title='Género'),
    tooltip=[
        alt.Tooltip('Década:O'),
        alt.Tooltip('Género:N'),
        alt.Tooltip('Cantidad:Q')
    ]
).properties(
    title='Evolución de los géneros nominados a Mejor Película por década',
    width=700,
    height=450
)

grafico_2
```

```python
grafico_2.save("grafico_2.html")
files.download("grafico_2.html")
```

```python
# Gráfico 3: Géneros ganadores a "Mejor Película" a lo largo de los años

import pandas as pd
import altair as alt

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("Base de datos Entrega 3.csv", sep=";")

# Filtrar solo películas ganadoras
ganadoras = df[df['LUGAR'] == 'Ganadora'].copy()

# Crear variable década
ganadoras['Década'] = (ganadoras['AÑO ESTRENO'] // 10) * 10

# Tabla de frecuencias por década y género
generos_decada = pd.crosstab(
    ganadoras['Década'],
    ganadoras['GÉNERO']
).reset_index()

# Pasar a formato largo
generos_decada_long = generos_decada.melt(
    id_vars='Década',
    var_name='Género',
    value_name='Cantidad'
)

# Eliminar combinaciones con 0 películas
generos_decada_long = generos_decada_long[
    generos_decada_long['Cantidad'] > 0
]

# Gráfico
grafico_3 = alt.Chart(generos_decada_long).mark_area().encode(
    x=alt.X('Década:O', title='Década'),
    y=alt.Y('Cantidad:Q', title='Cantidad de películas ganadoras'),
    color=alt.Color('Género:N', title='Género'),
    tooltip=[
        alt.Tooltip('Década:O'),
        alt.Tooltip('Género:N'),
        alt.Tooltip('Cantidad:Q')
    ]
).properties(
    title='Evolución de los géneros ganadores del Óscar a Mejor Película por década',
    width=700,
    height=450
)

grafico_3
```

```python
grafico_3.save("grafico_3.html")
files.download("grafico_3.html")
```
