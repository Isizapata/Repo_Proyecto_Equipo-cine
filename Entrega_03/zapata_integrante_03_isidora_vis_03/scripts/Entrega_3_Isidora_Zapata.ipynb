Un gráfico que muestre a los 20 directores con más peliculas nominada o ganadoras de los Premios Oscar. El amarillo para las películas ganadoras y el gris como nominadas.

```python

df = pd.read_csv('database premios.csv', sep=';')

# 1. Filtramos los años entre 1977 y 2026
df_filtrado = df[(df['AÑO ESTRENO'] >= 1977) & (df['AÑO ESTRENO'] <= 2026)].copy()

# 2. Seleccionamos los 20 directores con más actividad
top_20 = df_filtrado['DIRECTOR'].value_counts().nlargest(20).index.tolist()
df_top = df_filtrado[df_filtrado['DIRECTOR'].isin(top_20)]

# 3. Crear el gráfico de puntos interactivo
chart_puntos = alt.Chart(df_top).mark_circle(size=120, opacity=0.8).encode(
    # Eje X: Años (usamos :O para que se vea como escala ordinal/discreta)
    x=alt.X('AÑO ESTRENO:O',
            title='Año de Estreno'),

    # Eje Y: Directores
    y=alt.Y('DIRECTOR:N',
            title='Director',
            sort='-x'), # Ordenados por cantidad total

    # Color: Dorado para Ganadora, Plata para Nominada
    color=alt.Color('LUGAR:N',
                    title='Resultado',
                    scale=alt.Scale(domain=['Ganadora', 'Nominada'],
                                   range=['#FFD700', '#C0C0C0'])),
    # Tooltip: Información detallada al pasar el ratón
    tooltip=[
        alt.Tooltip('PELÍCULA :N', title='Película'),
        alt.Tooltip('AÑO ESTRENO:O', title='Año'),
        alt.Tooltip('DIRECTOR:N', title='Director'),
        alt.Tooltip('LUGAR:N', title='Resultado')
    ]
).properties(
    title='Top 20 directores con más peliculas: Ganadoras vs Nominadas a "Mejor Película" Premios Oscar (1970-2026)',
    width=800,
    height=450
).configure_axisY(
    grid=True,              # Activa las líneas horizontales
    gridColor='#EAEAEA',    # Color gris muy suave para no distraer
    gridDash=[4, 4],        # Hace que la línea sea punteada [punteado, espacio]
    gridWidth=1             # Grosor de la línea
    ).interactive() # Esto permite hacer zoom y mover el gráfico
# --- PASO 3: Mostrar el resultado ---
chart_puntos
```
