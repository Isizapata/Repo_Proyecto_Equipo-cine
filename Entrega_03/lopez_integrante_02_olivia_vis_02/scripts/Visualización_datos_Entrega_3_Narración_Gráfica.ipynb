```python
# Gráfico 2 Géneros de películas

import pandas as pd
import altair as alt

from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Base de datos Entrega 3.csv", sep=";")
df.head()

conteo = df["GÉNERO"].value_counts().reset_index()
conteo.columns = ["GÉNERO", "Cantidad"]

grafico_2 = alt.Chart(conteo).mark_bar().encode(
    x="GÉNERO:N",
    y="Cantidad:Q",
    color="GÉNERO:N",
    tooltip=["GÉNERO", "Cantidad"]
).properties(
    title="Cantidad de películas por género")

grafico_2
```

```python
# Gráfico 3 Evolución de los géneros cinematográficos

import pandas as pd
import altair as alt

from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Base de datos Entrega 3.csv", sep=";")
df.head()

df['DÉCADA'] = (df['AÑO ESTRENO'] // 10) * 10

conteo_decada = df.groupby(['DÉCADA', 'GÉNERO']).size().reset_index(name='Cantidad')

grafico_3 = alt.Chart(conteo_decada).mark_rect().encode(
    x=alt.X('DÉCADA:O', title='Década'),
    y=alt.Y('GÉNERO:N', title='Género'),
    color=alt.Color('Cantidad:Q', scale=alt.Scale(scheme='blues'), title='Nº Películas'),
    tooltip=['DÉCADA', 'GÉNERO', 'Cantidad']
).properties(
    title='Evolución de Géneros por Década',
    width=500,
    height=450
)

grafico_3

grafico_3.save("grafico_3.html")
!pip install vl-convert-python

grafico_3.save("grafico_3.png")

files.download("grafico_3.html")
#!pip install vl-convert-python
#files.download("grafico_3.png")
```

```python
# Gráfico 3 Géneros nominados vs ganadores a "Mejor Película"

import pandas as pd
import altair as alt

from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Base de datos Entrega 3.csv", sep=";")
df.head()

comparacion = pd.crosstab(df['GÉNERO'], df['LUGAR'])
comparacion = comparacion.reset_index()

comparacion = comparacion.sort_values('Ganadora', ascending=False).head(10)

comparacion_long = comparacion.melt(
    id_vars='GÉNERO',
    var_name='Resultado',
    value_name='Cantidad'
)

grafico_3 = alt.Chart(comparacion_long).mark_bar().encode(
    x=alt.X('GÉNERO:N', sort='-y', title='Género'),
    y=alt.Y('Cantidad:Q', title='Cantidad'),
    color=alt.Color('Resultado:N', title='Resultado'),
    xOffset='Resultado:N',
    tooltip=['GÉNERO', 'Resultado', 'Cantidad']
).properties(
    title='Nominadas vs Ganadoras por Género (Top 10)',
    width=600,
    height=400
)

grafico_3
```

```python
grafico_3.save("grafico_3.html")
files.download("grafico_3.html")
```
